```python
#1. House Sales in King County, USA

#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#Loading the dataset
df = pd.read_csv('kc_house_data.csv')

# .shape — returns (number of rows, number of columns)
print('Shape:', df.shape)

#preview first 5 rows
print(df.head())
```

```python
#summary of data types and non-null counts
print(df.info())
```

```python
#1. Data cleaning and handling missing values

#i. finding missing values

#isnull() returns a True value if it contains a missing value 
#sum() adds up the boolean values. 

missing = df.isnull().sum()
print('Missing values per column:')
print(missing)

```

```python
#ii. checking for duplicate rows
#count duplicate rows
print('Duplicate rows:', df.duplicated().sum())
#duplicated() evaluates every row and returns a boolean Series 
```

```python
#iii. Checking for incorrect or inconsistent values

# bedrooms: a house with 0 or 33 bedrooms is suspicious
print('Bedroom value counts (unusual values):')
print(df['bedrooms'].value_counts().sort_index())

# price: a price of 0 would be incorrect
print('\nHouses with price = 0:', (df['price'] == 0).sum())
```

```python
#simply drop suspicious data
# .drop() with a boolean condition removes rows where the condition is True
df = df[df['bedrooms'] <= 15]   # keep only houses with 15 or fewer bedrooms
print('Shape after removing outlier bedroom row:', df.shape)
```

```python
# pd.to_datetime() — converts a string to a real date
df['date'] = pd.to_datetime(df['date'])

# Drop columns we won't use in analysis (id is just an identifier; lat/long
# are geographic coordinates not needed for EDA)
df.drop(columns=['id', 'lat', 'long'], inplace=True)

print('Cleaned dataset shape:', df.shape)
print('Date column type:', df['date'].dtype)
```

```python
#Descriptive Statistics
# .describe() — automatically calculates count, mean, std, min, max,
# and quartiles (25%, 50%, 75%) for every numeric column
df.describe().round(2)
```

```python
# Detailed stats for the 'price' column — the most important variable

price = df['price']

print('=== Price Statistics ===')
print(f"Mean      : ${price.mean():,.0f}")         # average price
print(f"Median    : ${price.median():,.0f}")       # middle value (less affected by extremes)
print(f"Mode      : ${price.mode()[0]:,.0f}")      # most frequent price
print(f"Std Dev   : ${price.std():,.0f}")          # how spread out prices are
print(f"Variance  : ${price.var():,.0f}")          # std dev squared
print(f"Min       : ${price.min():,.0f}")          # cheapest house
print(f"Max       : ${price.max():,.0f}")          # most expensive house
print(f"Q1 (25%)  : ${price.quantile(0.25):,.0f}")  # 25th percentile
print(f"Q3 (75%)  : ${price.quantile(0.75):,.0f}")  # 75th percentile
print(f"IQR       : ${price.quantile(0.75) - price.quantile(0.25):,.0f}")  # interquartile range

# Interpretation: The mean is higher than the median, which means
# a few very expensive houses are pulling the average up (right-skewed data).
```

```python
#Data Visualization

# A histogram shows how frequently values fall into equal-width bins.
# It helps us understand the shape of the price distribution.

plt.figure(figsize=(10, 5))    # set the canvas size (width, height) in inches

# plt.hist() — draws the histogram
# bins=50 means we divide the range into 50 buckets
# edgecolor adds a thin border around each bar for clarity
plt.hist(df['price'], bins=50, color='steelblue', edgecolor='white')

plt.title('Distribution of House Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Number of Houses')

plt.tight_layout()   # prevents labels from being clipped
plt.show()

# Observation: The distribution is right-skewed — most houses are priced
# below $1M, but a few luxury homes push the tail far to the right.
```

```python
# A bar chart compares a numeric value across different categories.
# Here we compare average price for each bedroom count.

# .groupby() — groups rows by a column value
# ['price'].mean() — then calculates the average price within each group
avg_price_by_bed = df.groupby('bedrooms')['price'].mean().reset_index()

plt.figure(figsize=(10, 5))

# sns.barplot() — draws a bar chart; x=categories, y=values
sns.barplot(data=avg_price_by_bed, x='bedrooms', y='price', color='steelblue')

plt.title('Average House Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price (USD)')
plt.tight_layout()
plt.show()

# Observation: Prices generally increase from 1 to 5–6 bedrooms,
# then become inconsistent due to fewer data points at the high end.
```

```python
# A box plot shows the median, quartiles, and outliers of a distribution.
# It is useful for comparing two groups side by side.

plt.figure(figsize=(8, 5))

# sns.boxplot() — draws a box plot
# hue splits the boxes into groups (0 = no waterfront, 1 = waterfront)
sns.boxplot(data=df, x='waterfront', y='price')

plt.title('House Price: Waterfront vs Non-Waterfront')
plt.xlabel('Waterfront (0 = No, 1 = Yes)')
plt.ylabel('Price (USD)')
plt.tight_layout()
plt.show()

# Observation: Waterfront houses have a noticeably higher median price
# and more spread — confirming that water views command a premium.
```

```python
# A scatter plot shows the relationship between two numeric variables.
# Each dot represents one house.

plt.figure(figsize=(9, 5))

# plt.scatter() — plots individual points at (x, y) coordinates
# alpha controls transparency so overlapping points are visible
plt.scatter(df['sqft_living'], df['price'], alpha=0.2, color='steelblue', s=5)

plt.title('Living Area vs House Price')
plt.xlabel('Living Area (sqft)')
plt.ylabel('Price (USD)')
plt.tight_layout()
plt.show()

# Observation: There is a clear positive relationship — larger houses
# tend to cost more. The relationship appears roughly linear.
```

```python
# A correlation heatmap shows how strongly pairs of numeric variables
# are related. Values range from -1 (opposite) to +1 (perfectly aligned).

# Select a subset of meaningful numeric columns for readability
cols = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
        'sqft_lot', 'floors', 'waterfront', 'view', 'condition',
        'grade', 'yr_built']

# .corr() — calculates the correlation coefficient between every pair of columns
corr_matrix = df[cols].corr()

plt.figure(figsize=(11, 8))

# sns.heatmap() — draws a color-coded grid of correlation values
# annot=True prints the number inside each cell
# fmt='.2f' formats numbers to 2 decimal places
# cmap chooses the color scheme
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Heatmap of House Features')
plt.tight_layout()
plt.show()

# Observation: 'sqft_living' and 'grade' have the strongest positive
# correlations with price (0.70 and 0.67 respectively), meaning bigger
# houses and higher build quality are the top price drivers.
```

```python
#Hypothesis Formulation and Statistical Testing

# --- Research Question ---
# Do waterfront properties sell for a significantly higher price
# than non-waterfront properties?
#
# H₀ (Null Hypothesis)     : The average price of waterfront and
#                             non-waterfront houses is the same.
# H₁ (Alternative Hypothesis): Waterfront houses have a higher
#                              average price than non-waterfront houses.
#
# Test chosen: Independent samples t-test
# Reason: We are comparing the means of two separate, independent groups
# (waterfront vs non-waterfront). A t-test is the standard tool for this.

# Separate the two groups
waterfront_prices     = df[df['waterfront'] == 1]['price']
non_waterfront_prices = df[df['waterfront'] == 0]['price']

print(f"Waterfront houses     : {len(waterfront_prices)} houses, avg price ${waterfront_prices.mean():,.0f}")
print(f"Non-waterfront houses : {len(non_waterfront_prices)} houses, avg price ${non_waterfront_prices.mean():,.0f}")
```

```python
# stats.ttest_ind() — performs an independent samples t-test
# It returns:
#   t_stat  : how many standard errors the group means are apart
#   p_value : probability of seeing this difference if H₀ were true
#
# equal_var=False uses Welch's version, which works even when
# the two groups have different variances (safer default choice)
t_stat, p_value = stats.ttest_ind(waterfront_prices, non_waterfront_prices,
                                   equal_var=False)

print(f"t-statistic : {t_stat:.4f}")
print(f"p-value     : {p_value:.6f}")
print()

# Alpha (significance level) = 0.05 — the standard threshold.
# If p < 0.05 we reject H₀ (the difference is statistically significant).
alpha = 0.05
if p_value < alpha:
    print('Result: Reject H₀ — waterfront houses have a statistically significantly higher price.')
else:
    print('Result: Fail to reject H₀ — no significant price difference found.')
```

```python
#Interpretation and Conclusions

# 1. DATA QUALITY
#    - The dataset is clean: no missing values and no duplicate rows.
#    - One row with 33 bedrooms was removed as a likely data entry error.

# 2. PRICE DISTRIBUTION
#    - House prices are right-skewed: most homes fall below $1M,
#      but a small number of luxury properties push the mean above the median.
#    - Mean price  ≈ $540,000, Median price ≈ $450,000

# 3. KEY PRICE DRIVERS (from correlation heatmap)
#    - sqft_living (living area) has the strongest correlation with price (r ≈ 0.70).
#    - grade (construction quality) is the second strongest driver (r ≈ 0.67).
#    - Bathrooms and sqft_above also correlate positively with price.

# 4. WATERFRONT PREMIUM
#    - Waterfront homes average ~$1,662,524 vs ~$531,563 for non-waterfront.
#    - The t-test confirms this difference is statistically significant (p < 0.05),
#      so we reject the null hypothesis.
#    - Waterfront location adds roughly $1.1M on average to a home's price.

# Conclusion 
# Living area and build grade are the most reliable predictors of house price
# in King County. Waterfront access represents a substantial and statistically
# proven premium. Buyers and sellers should consider these factors when
# evaluating property values in this market.
```
